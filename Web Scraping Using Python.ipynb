```python
table = soup.find('table')

table
```

```python
t_row = table.find_all('tr')

t_row
```

```python
data = []

for row in table.find_all('tr'):
    row_data = []
    for cell in row.find_all('th'):
        row_data.append(cell.text)
    
    for cell in row.find_all('td'):
        row_data.append(cell.text)
    data.append(row_data)
```

```python
import pandas as pd

pd.DataFrame(data)
```
